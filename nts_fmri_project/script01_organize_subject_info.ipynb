```python
import pandas as pd

df_questionnaire = pd.read_csv('CC_RC_Master_Database_10.23.18_Data.csv')
df_diagnosis     = pd.read_csv('nts_and_tis_mdd_and_ptsd.csv')

df_questionnaire = df_questionnaire[['CC_Pre_Intake_ID', 'CC_Post_Intake_ID', 
                                     'GAD_7', 'Current_GAD', 'Lifetime_GAD', 
                                     'Criterion_A_Trauma',  
                                     'Current_PTSD_Diagnosis', 'Lifetime_PTSD_Diagnosis', 
                                     'Current_MDD', 'Lifetime_MDD', 
                                     'OCD', 'Social_Phobias', 
                                     'Other_Phobias', 'Panic_Disorder',
                                     'PHQ_Score_Screen', 'PHQ_Score_Intake', 'PHQ_Score_mean']]

df_questionnaire
```

```python

df_diagnosis['Record ID:'] = df_diagnosis['Record ID:'].astype(str)
df_diagnosis['Record ID:'] = df_diagnosis['Record ID:'].str[-4:].astype(int)
```

```python
# check mdd diagnosis:

df_merge = df_questionnaire.merge(df_diagnosis, left_on = 'CC_Pre_Intake_ID',
                                 right_on = 'Record ID:').drop('Record ID:', axis = 1)

df_merge['PHQ1>=10'] = df_merge['PHQ_Score_Screen'] >= 10
df_merge['PHQ1>=10'] = df_merge['PHQ1>=10'].astype(int)

df_merge['PHQ2>=10'] = df_merge['PHQ_Score_Intake'] >= 10
df_merge['PHQ2>=10'] = df_merge['PHQ2>=10'].astype(int)

# show rows that current MDD or life time MDD are not correct

# df_merge.loc[(df_merge['PHQ1>=10'] != df_merge['Current MDD']) |\
#              (df_merge['PHQ2>=10'] != df_merge['Lifetime MDD']), :].head()

df_merge.loc[(df_merge['PHQ1>=10'] != df_merge['Current MDD']) |\
             (df_merge['PHQ2>=10'] != df_merge['Lifetime MDD']), :]
```

```python
import glob

base_dir = "/Users/jingjiang/Library/CloudStorage/GoogleDrive-jingjiangpsy@gmail.com/Shared drives/Jing_Drive/"
img_dir1 = "3_Data_Working/CausalConnectome/post_processing/tmsfmri/group_analysis_HRF_all/L_pMFG/NTS_2ndlevel/"
img_dir2 = "3_Data_Working/CausalConnectome/post_processing/tmsfmri/group_analysis_HRF_all/L_aMFG/NTS_2ndlevel/"
img_dir3 = "3_Data_Working/CausalConnectome/post_processing/tmsfmri/group_analysis_HRF_all/R_M1/NTS_2ndlevel/"

img_file_nts = glob.glob(base_dir + img_dir1 + "CausCon_NTS*.nii") 
    
# img_file_nts = glob.glob(base_dir + img_dir1 + "CausCon_NTS*.nii") + \
#                glob.glob(base_dir + img_dir2 + "CausCon_NTS*.nii") + \
#                glob.glob(base_dir + img_dir3 + "CausCon_NTS*.nii")


file_name = [x.split('/')[-1] for x in img_file_nts]
subject_id = [int(x.split('_')[1][-4:]) for x in file_name]
site = [x.split('_')[3] + '_' + x.split('_')[4] for x in file_name]

df_img_file = pd.DataFrame({'file_dir': img_file_nts, 
                            'file_name': file_name,
                            'site': site,
                            'subject_id': subject_id})

df_img_file.head()
```

```python
from IPython.display import display
# remove selected disorders:

df_img_file_merge = df_img_file.merge(df_questionnaire, left_on = 'subject_id', right_on = 'CC_Post_Intake_ID')

remove_disorder = ['Criterion_A_Trauma',  
                   'Current_PTSD_Diagnosis', 'Lifetime_PTSD_Diagnosis', 
                   #'Current_MDD', 'Lifetime_MDD', 
                   'Current_GAD', 'Lifetime_GAD', 
                   'OCD', 'Social_Phobias', 
                   'Other_Phobias', 'Panic_Disorder']

# df_img_file_merge = df_img_file_merge.loc[(df_img_file_merge[remove_disorder].sum(axis = 1) == 0), :]
# df_img_file_merge

# df_img_file_merge = df_img_file_merge.loc[(df_img_file_merge[remove_disorder].sum(axis = 1) == 0) &
#                                          ((df_img_file_merge['Current_MDD'] == 1) | 
#                                           (df_img_file_merge['Lifetime_MDD'] == 1))
#                                           , :]

df_merge_life_time_mdd = df_img_file_merge.loc[(df_img_file_merge[remove_disorder].sum(axis = 1) == 0) &
                                          (df_img_file_merge['Lifetime_MDD'] == 1)
                                          , :]

df_merge_current_mdd = df_img_file_merge.loc[(df_img_file_merge[remove_disorder].sum(axis = 1) == 0) &
                                          (df_img_file_merge['Current_MDD'] == 1)
                                          , :]

df_merge_life_time_mdd.to_csv('out01_img_file_info_life_time_mdd.csv', index = None)
print(df_merge_life_time_mdd.shape)
display(df_merge_life_time_mdd)

df_merge_current_mdd.to_csv('out01_img_file_info_current_mdd.csv', index = None)
print(df_merge_current_mdd.shape)
df_merge_current_mdd
```

```python
## using PHQ >= 10 and GAD <= 10 to select mdd:

df_img_file_merge = df_img_file.merge(df_questionnaire, left_on = 'subject_id', right_on = 'CC_Post_Intake_ID')


df_img_file_merge = df_img_file_merge.loc[(df_img_file_merge['PHQ_Score_Intake'] >= 10) \
                                        & (df_img_file_merge['GAD_7'] <= 10), :]

df_img_file_merge.to_csv('out01_img_file_info_phq.csv', index = None)
print(df_img_file_merge.shape)
df_img_file_merge
```
