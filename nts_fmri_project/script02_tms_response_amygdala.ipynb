```python
# extract tms response in the amygdala:
# using images generated in last step:

import pandas as pd

## NTS group: use file created in last step:

input_file = "out01_img_file_info_life_time_mdd.csv"
output_file = "out02_img_file_info_mdd_tms_response.csv"

# input_file = "out01_img_file_info_current_mdd.csv"
# output_file = "out02_img_file_info_current_mdd_tms_response.csv"

# input_file = "out01_img_file_info_phq.csv"
# output_file = "out02_img_file_info_phq_tms_response.csv"

data = pd.read_csv(input_file)

data.head()
```

```python
# list roi mask:
import glob
from nilearn import maskers
import numpy as np

# FSL_R_Amy_thr40
# ROI_R_Amy_Com_MNI
# FIRST_R_amyg_small
# ROI_B_Amy_Com_MNI
# FSL_L_Amy_thr40
# FIRST_B_amyg_small
# FSL_B_Amy_thr40
# ROI_L_Amy_Com_MNI
# FIRST_L_amyg_small

# base_dir = "/Users/jingjiang/Library/CloudStorage/GoogleDrive-jingjiangpsy@gmail.com/Shared drives/Jing_Drive/"
# roi_dir = "3_Data_Working/CausalConnectome/post_processing/tmsfmri/ROI_masks/Amy/"

base_dir = "/home/niki/Downloads/JiangLab/"
roi_dir = "target_mask/"


mask_dir = glob.glob(base_dir + roi_dir + "*.nii")
mask_name = [x.split('/')[-1][:-4] for x in mask_dir]

for m, m_name in zip(mask_dir, mask_name):
    print(m_name)
    tms_site_response = []
    
    for i in range(data.shape[0]):
        image = data.iloc[i, 0]
        nifti_masker = maskers.NiftiMasker(m).fit()
        tms_site_response.append(np.mean(nifti_masker.transform(image)))
        
    data[m_name] = tms_site_response
```

```python
data.to_csv(output_file, index = None)
```
